## Numerical Interpolation

* Approximating functions, given you know some of the points.


<figure>
    <img src="interpolating ground state.png" width=1000>
</figure>

Imagine you know the spatial probability density of a particle at 10 places within a different well.

How can you estimate the proability density in between these points?

<figure>
<img src="ground state interpolation schemes.png">
    <figcaption>
        Interpolating the probability density using a) nearest neighbour b) linear interpolation c) higher order interpolation
    </figcaption>
</figure>

There's a handy package that can do this for you: `scipy.interpolate.interp1d`

Documentation: https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html

```python
import numpy as np

def groundstate_prob_density(x):
    return 2*np.sin(np.pi*x)**2

x = np.array([0, 0.15, 0.2, 0.22, 0.4, 0.65, 0.7, 0.8, 0.88, 1])
p = groundstate_prob_density(x)
```

```python
### KNOWN POINTS:
# print(x)
# print(p)

### PLOT (PROPERLY) to see what they look like (explain args for figure and add_subplot)
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('$x$', fontsize=15)
ax.set_ylabel('$P(x)$', fontsize=15)
ax.grid()
ax.plot(x, p, markersize=10, marker='o', color='darkblue', linestyle='None')

plt.show()

```

```python
### IMPORT INTERP1D
from scipy.interpolate import interp1d
interpolatingFunc = interp1d(x, p, kind='quadratic')
xAll = np.linspace(0, 1, 1000)
pAll = interpolatingFunc(xAll)
### TRY NEAREST NEIGHBOUR, LINEAR, CUBIC
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1,1,1)
ax.plot(x, p, markersize=10, color='darkblue', marker='o', linestyle='None')
ax.plot(xAll, pAll, color='darkgreen')
ax.set_ylabel('$P(x)$', fontsize=15)
ax.set_xlabel('$x$', fontsize=15)
ax.set_xlim(0, 1)
ax.grid()
plt.show()

```

## Numerical Integration

* Evaluating definite integrals: $\int_a^b f(x) dx$ - typically where you can't solve them analytically <font size=2>(that is with pen and paper)</font>


* Running with the infinite potential example:

What is the probability of a particle being within an interval $x\in [a,b]$?

$$ P(a<x<b) = \int_{a}^{b} \psi(x)^* \psi(x) dx $$

This is an integral we can use numerical integration to evaluate for given $a$, $b$

In 1D, integrals are essentially just 'areas' under curves.

So how should we approximate this area?

Using `integrate.quad` with an easy example: evaluate $\int_0^1 x^2 dx = \frac{1}{3}\left[x^3\right]^1_0 = \frac{1}{3} $

```python
#### integrate.quad - easy example
def quadratic(x):
    return x**2

print(integrate.quad(quadratic, 0, 1))




```

Now use `integrate.quad` to do something more exciting: probability of being on the left of the well:
$$ P(0<x<0.5) = \int_0^{0.5} 2 sin^2(\pi x) dx $$

```python
### Define the function
def prob_density(x):
    return 2*np.sin(np.pi*x)**2

probability = integrate.quad(prob_density, 0, 0.25)

print(probability)


```

`integrate.trapezoidal` and `integrate.simpson` work similarly: they return a VALUE, with an error. HOWEVER, they take arrays of points rather than a specified FUNCTION.

```python
### USE GROUND STATE PROB DENSITY AS EXAMPLE (plot as a reminder)

x = np.linspace(0, 0.5, 10)
p = prob_density(x)

probability = integrate.trapezoid(p, x)
print(probability)
```

### EVERYONES FAVOURITE: CUMTRAPZ!!!

* `scipy.integrate.cumulative_trapezoid` <font size=2>previously known as `cumtrapz`</font> is a method that can be used for evaluating indefinite integrals. e.g.: 
$$ \int x dx = \frac{1}{2} x^2 + C $$

* Cumulatively 'keeps track' of areas under a given curve.

```python
### DEFINE A FUNCTION TO GET A SET OF POINTS
def linear(x):
    return x

x = np.linspace(0, 3, 10)
y = linear(x)

res = integrate.cumulative_trapezoid(y, x, initial=0)

print(x, '\n', res)

### WHAT IS THE OUTPUT?
plt.plot(x, y, color='orange')
plt.plot(x, res, color='darkgreen')
```
