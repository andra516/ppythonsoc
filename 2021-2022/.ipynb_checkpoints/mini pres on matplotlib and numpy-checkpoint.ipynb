### MULTI-FIGURE PLOTS




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##### CREATE THE GRID

gridSize = 500

dX = 3/gridSize
dY = 3/gridSize

realMat = np.zeros(shape=(gridSize,gridSize))
imagMat = np.zeros(shape=(gridSize, gridSize))

for k in range(gridSize):
    realMat[:,k] = -2 + dX*k
    imagMat[k,:] = -1.5 + dY*k
    
cMat = realMat + 1j* imagMat

print(cMat.flatten())
```

```python
def apply_rule(c):
    nIters = 10
    
    z = c
    
    for _ in range(nIters):
        z = z**2 + c
        
    if np.abs(z) > 3:
        return False
    
    else:
        return True
```

```python
### ITERATE THROUGH EACH c AND APPLY RULE TO EACH

C = []
for c in cMat.flatten():
    if apply_rule(c):
        C.append(c)
    else:
        continue

```

```python

##### PLOT TO CHECK
realC = np.real(C)
imagC = np.imag(C)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

ax.scatter(realC, imagC, color='k', marker='.')
ax.set_xlim()

plt.show()


```
